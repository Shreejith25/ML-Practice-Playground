```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
iris.data
```

```python
iris.target
```

```python
iris.target_names
```

```python
# independent and dependent features

x = iris.data
```

```python
y = iris.target
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
## For post Pruning
tree_model =DecisionTreeClassifier()
```

```python
tree_model.fit(x_train,y_train)
```

```python
from sklearn import tree # to get image of the split and parameter value in each split
plt.figure(figsize=(15,10))
tree.plot_tree(tree_model,filled=True)
```

- Gini apart from initial node oscillates from 0 to 0.5 
- Value[] is the split between three catagories of target  in each stage 
- x[] shows the column or feature used to split

\* at level 2 i.e value is [0,32,1] and [0,5,36] we can prune as after the that the split is not necessary because the classification can be done easily it then with least error.

# Post Pruning

```python
model = DecisionTreeClassifier(max_depth=2,random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(10,10))
tree.plot_tree(model,filled=True)
```

```python
# Prediction stage

y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report

score = accuracy_score(y_pred,y_test)
print(f' Prediction Score after Post Pruning DT {score*100} %')
```

```python
report = classification_report(y_pred,y_test)

print(report)
```

# Pre Prunining

```python



parameters = {
    'criterion': ['gini', 'entropy', 'log_loss'],   
    'splitter': ['best', 'random'],                 
    'max_depth': [1, 2, 3, 4, 5],                  
    'max_features': ['auto', 'sqrt', 'log2'],  
        
    
}
```

```python
from sklearn.model_selection import GridSearchCV

```

```python
model1 = DecisionTreeClassifier()
cv = GridSearchCV(model1,param_grid=parameters,cv =5,scoring='accuracy')

# cv = 5 devides in 5 sets and makes one set as test data and rest as training 
#It repeats it untill all 5 sets are used as test data 
# and it will show which combination is best in accuracy
```

```python
cv.fit(x_train,y_train)
```

```python
cv.best_params_
```

```python
y_pred = cv.predict(x_test)
```

```python
score = accuracy_score(y_pred,y_test)
score
```

```python
rep = classification_report(y_pred,y_test)

print(rep)
```
